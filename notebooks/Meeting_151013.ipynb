#** E Field = 0.5 MV/m **

**Parallel Fields 1mm distance - Red Skewnormal fit- Green example gauss**

```python
rid=4864
par=False
par_str='Par_' if par else 'Anti_'
mean,stdev,dip_pos,dip_str = srkdata.get_data_for_rids_from_database([rid],par_str+"PhiMean,"+par_str+"PhiStDev,DipolePositionBelowChamber,DipoleFieldStrength")[0]
drawString="(phi-"+str(mean)+")/"+str(stdev)
c1=srkgraphics.make_root_plot_from_results_file(rid,par,drawString,"abs("+drawString+") < 5")
htemp = gPad.GetPrimitive("htemp")
htemp.GetXaxis().SetRangeUser(-5.,5.)
htemp.SetTitle("Normalized Phi - "+par_str+"Fields with Dipole "+str(dip_str)+" T m^{3},  "+str(-dip_pos*100.)+" cm Below Chamber;Normalized Phi;Counts")
func = TF1("skewnormal","[3]*TMath::Gaus((x-[1])/[2],0,1)*(1+TMath::Erf([0]*(x-[1])/[2]))",-5,5)
func.SetParNames("Skew Factor","Offset","Sigma","Amplitude")
func.SetParameters(1.5,-1.2,1.5,216)
func.SetParLimits(0,-10,10)
func.SetParLimits(1,-3,3)
func.SetParLimits(2,0.5,2)
func.SetParLimits(3,htemp.Integral()*0.0001,htemp.Integral())
htemp.Fit("skewnormal","M")
gaus = TF1("gaus","400*TMath::Gaus(x,0,1)",-5,5)
gaus.SetLineColor(ROOT.kGreen)
gaus.Draw("same")
func.Draw("same")
c1
```

** Fit this time to a Tsallis function(Red)  and a manual Gauss (Green)**

```python
rid=575
par=True
par_str='Par_' if par else 'Anti_'
mean,stdev,dip_pos,dip_str = srkdata.get_data_for_rids_from_database([rid],par_str+"PhiMean,"+par_str+"PhiStDev,DipolePositionBelowChamber,DipoleFieldStrength")[0]
drawString="(phi-"+str(mean)+")/"+str(stdev)
c1=srkgraphics.make_root_plot_from_results_file(rid,par,drawString,"abs("+drawString+") < 5")
htemp = gPad.GetPrimitive("htemp")
htemp.GetXaxis().SetRangeUser(-5.,5.)
htemp.SetTitle("Normalized Phi - "+par_str+"Fields with Dipole "+str(dip_str)+" T m^{3},  "+str(-dip_pos*100.)+" cm Below Chamber;Normalized Phi;Counts")
phiTsallisFunc = TF1("phiTsallisFunc", "[0]/pow(1+((x-[3])/[1])*((x-[3])/[1]),[2])", -5)
phiTsallisFunc.SetParNames("Skew Factor","Offset","Sigma","Amplitude")
phiTsallisFunc.SetParNames("Amplitude", "Sigma", "Power", "Mean");
phiTsallisFunc.SetParameters(30000, stdev * 0.005, .76, 0);
phiTsallisFunc.SetParLimits(3, 0, 0);
htemp.Fit("phiTsallisFunc","M")
phiTsallisFunc.Draw("same")
gaus = TF1("gaus","30000*TMath::Gaus(x,0,1)",-5,5)
gaus.SetLineColor(ROOT.kGreen)
gaus.Draw("same")
c1.SetLogy(0)
c1
```
